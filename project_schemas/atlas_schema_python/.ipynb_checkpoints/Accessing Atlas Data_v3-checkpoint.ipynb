# Download Images from S3 to Local Computer
(Relies on `bucket` and `filename_list` from previous cell)

```python
from subprocess import call

# S3 bucket must be appended to the s3_fp
def download_from_s3(local_root, s3_fp, raw=False):
    if raw:
        s3_url = "s3://mousebrainatlas-rawdata-backup/" + s3_fp
    else:
        s3_url = "s3://mousebrainatlas-data/" + s3_fp
    local_download_fp = local_root + s3_fp
    
    print("aws "+"s3 "+"cp "+s3_url+' '+local_download_fp)

    call(["aws",\
          "s3",\
          "cp",\
          s3_url,\
          local_download_fp])
```

```python
# Replace this with the fp you want to download to
local_root_alex_pc = "/mnt/c/Users/Alex/Documents/TEST/"
```

```python
# DOWNLOAD PROCESSED FILE
MD585_slice_200 =  MD585_slices[200]
s3_fp = MD585_slice_200['processed_s3_fp'] 


download_from_s3( local_root_alex_pc, s3_fp, raw=False)
```

```python
# DOWNLOAD PROCESSED FILE
MD585_slice_200 =  MD585_slices[200]
s3_fp = MD585_slice_200['raw_s3_fp'] 


download_from_s3( local_root_alex_pc, s3_fp, raw=True)
```
